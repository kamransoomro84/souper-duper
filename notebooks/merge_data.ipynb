```python
import pandas
from datetime import datetime

pandas.options.display.max_rows = None
```

```python
def convert_start_date(row):
    result = datetime.strptime(
        '{}/{}/{}'.format(
            row['start_day'],
            row['start_month'],
            row['start_year']
        ),
        '%d/%m/%Y'
    )
    return result

def convert_end_date(row):
    result = datetime.strptime(
        '{}/{}/{}'.format(
            row['end_day'],
            row['end_month'],
            row['end_year']
        ),
        '%d/%m/%Y'
    )
    return result

def add_event(row):
        events = date_frame[(date_frame['start_date'] <= row['date']) & (row['date'] <= date_frame['end_date'])]
        return 1 if events.shape[0]
```

```python
date_frame = pandas.read_csv('../data/event_data.csv')
date_frame['start_date'] = date_frame.apply(convert_start_date,axis=1)
date_frame['end_date'] = date_frame.apply(convert_end_date,axis=1)
date_frame
```
